#####ºEn la teoria de Markowitz es un objeto determinista, maximizando rendimientos esperados y pones una condicion a la volatilidad

```python
#Vamos a calcular la volatilidad de los rendimientos
volatilidad=apple["Adj Close"].pct_change().std()
volatilidad
```

```python
volatilidad_total=volatilidad*math.sqrt(252) #tambien se puede elevar a 253**0.5
volatilidad_total
```
