```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
#Escogimos 30 tickers de la Bolsa Mexicana de Valores
tickers = ['AMXB.MX', 'GMEXICOB.MX', 'FEMSAUBD.MX', 'TLEVISACPO.MX', 'CEMEXCPO.MX',
               'WALMEX.MX', 'BIMBOA.MX', 'GFINBURO.MX', 'GFNORTEO.MX', 'KOFUBL.MX',
               'ALFAA.MX', 'GRUMAB.MX', 'ELEKTRA.MX', 'GAPB.MX', 'LABB.MX',
               'NEMAKA.MX', 'CUERVO.MX', 'AAPL.MX', 'ALSEA.MX', 'GENTERA.MX',
               'OMAB.MX', 'RA.MX', 'PINFRA.MX', 'GCC.MX', 'LIVEPOLC-1.MX',
               'AC.MX', 'MEGACPO.MX', 'WALMEX.MX', 'GCARSOA1.MX', "Q.MX", "VESTA.MX"]
precios=yf.download(tickers,start="2015-12-31")["Close"]
precios
```

```python
#Sacamos lo rendimientos de cada uno de los tickers
rendimientos=precios.pct_change().dropna()
rendimientos
```

```python
#Sacamos la varianza de los rendimientos
varianza=rendimientos.var()
varianza
```

```python
#Sacamos de promedio de los rendimientos
mean=rendimientos.mean()
mean
```

```python
#Importamos un modulo para sacar la grÃ¡fica de los clusters
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python

# Crear un nuevo DataFrame con rendimientos y varianza
df_selected = pd.DataFrame({'Returns': mean, 'Variance': varianza})

# Normalizar los datos para que tengan la misma escala
scaler = StandardScaler()
df_normalized = scaler.fit_transform(df_selected)

# Aplicar el algoritmo de k-means
kmeans = KMeans(n_clusters=3, random_state=42)
df_selected['Cluster'] = kmeans.fit_predict(df_normalized)

# Visualizar los resultados (grÃ¡fica)
plt.scatter(df_selected['Returns'], df_selected['Variance'], c=df_selected['Cluster'], cmap='viridis')
plt.title('Clustering Financiero')
plt.xlabel('Rendimientos')
plt.ylabel('Varianza')
plt.show()
```
