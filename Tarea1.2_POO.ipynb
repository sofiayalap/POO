```python
#Ejercicio 4
def transform_column(column):
  mean=column.mean()
  column = column - mean
  std_dev = column.std()
  column = column / std_dev
  return column
```

```python
dr["AAPL"]=transform_column(dr['AAPL'])
dr["AAPL"]
```

```python
#Ejercicio 5
dr["JPM"]=transform_column(dr['JPM'])
dr["JPM"]
```

```python
dr["WMT"]=transform_column(dr['WMT'])
dr["WMT"]
```

```python
dr["TGT"]=transform_column(dr['TGT'])
dr["TGT"]
```

```python
dr["MSFT"]=transform_column(dr['MSFT'])
dr["MSFT"]
```

```python
dr["AMGN"]=transform_column(dr['AMGN'])
dr["AMGN"]
```
